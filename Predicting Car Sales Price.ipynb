## Univariate Model Function with k=5 (*default*)

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
' Write a function that trains and test univariate models using the default k = 5 '

def knn_train_test(training_column,target_column,df):
    # RANDOMIZE THE ORDER TO AVOID BIASEDNESS 
    np.random.seed(1)
    
    # Randomize order of rows in data frame.
    shuffled_index = np.random.permutation(df.index)
    random_df = df.reindex(shuffled_index)
    
    # Initiate the class 
    knn = KNeighborsRegressor(algorithm = 'brute',n_neighbors = 5)
    
    # Split the dataframe in half 
    last_train_row = int(len(random_df) / 2)
    
    # Select the first half and set as training set.
    # Select the second half and set as test set.
    train_df = random_df.iloc[0:last_train_row]
    test_df = random_df.iloc[last_train_row:]
    
    # Fit the model 
    knn.fit(train_df[[training_column]],train_df[target_column])
    
    # Predict
    predictions = knn.predict(test_df[[training_column]])
    
    # Calculate mse 
    mse = mean_squared_error(test_df[target_column], predictions) 
    
    # Calculate rmse
    rmse = np.sqrt(mse)

    return(rmse)
    
```

```python
' Try out the function above using the "height" attribute '

knn_train_test(training_column ="height",target_column= "price",df = numeric_cars)
```

```python
# Drop the price column in the iteration list since it will be our target 
train_cols = numeric_cars.columns.drop('price')
rmse_val = dict()
    
for cols in train_cols:
    result = knn_train_test(training_column = cols, target_column = 'price', df = numeric_cars)
    rmse_val[cols]= result  

print(rmse_val)    
```

## Univariate Model Function on a Single Column for Different  *k* values

```python
' Write a function that trains and test univariate models using different values of k' 

def knn_train_test(training_column, target_column, df):
    np.random.seed(1)
        
    # Randomize order of rows in data frame.
    shuffled_index = np.random.permutation(df.index)
    random_df = df.reindex(shuffled_index)

    # Divide number of rows in half and round.
    last_train_row = int(len(random_df) / 2)
    
    # Select the first half and set as training set.
    # Select the second half and set as test set.
    train_df = random_df.iloc[0:last_train_row]
    test_df = random_df.iloc[last_train_row:]
    
    k_values = [1,3,5,7,9]
    k_rmses = {}
    
    for k in k_values:
        # Fit model using k nearest neighbors.
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[[training_column]], train_df[target_column])

        # Make predictions using model.
        predicted_labels = knn.predict(test_df[[training_column]])

        # Calculate and return RMSE.
        mse = mean_squared_error(test_df[target_column], predicted_labels)
        rmse = np.sqrt(mse)
        
        k_rmses[k] = rmse
    return k_rmses

```

```python
' Trying diferent k values on "height" attribute '  
knn_train_test(training_column = 'height',target_column ='price',df = numeric_cars)
```

## Univariate Model Function on all Columns for Different *k* values

```python
# Drop the price column in the iteration list since it will be our target 
train_cols = numeric_cars.columns.drop('price')
k_rmses_all_columns = dict()
    
for cols in train_cols:
    result = knn_train_test(training_column = cols, target_column = 'price', df = numeric_cars)
    k_rmses_all_columns[cols]= result   
    
k_rmses_all_columns
```

## Compute average RMSE of different *k* values for each feature. 

```python
# Calculate the average k value RMSE's for each feature and store them in a
# dictionary
avg_k_features_rmse = {}
for k,v in k_rmses_all_columns.items():
    avg = np.mean(list(v.values()))
    avg_k_features_rmse[k] =avg 
    
# Turn the Dictionary to a Series
series_avg_k_features_rmse = pd.Series(avg_k_features_rmse)

# Sort the values from lowest to highest
sorted_avg_k_features_rmse = series_avg_k_features_rmse.sort_values(ascending = True)

# Extract the sorted values Index
sorted_features = sorted_avg_k_features_rmse.index
print(sorted_features)

print(sorted_avg_k_features_rmse )
```

## Multivariate Model Function with k = 5 (*default*)

```python
' Write a function that trains and test multivariate models using the default k=5 '

def knn_train_test(training_columns, target_column, df):
    np.random.seed(1)
    
    # Randomize order of rows in data frame.
    shuffled_index = np.random.permutation(df.index)
    random_df = df.reindex(shuffled_index)

    # Divide number of rows in half and round.
    last_train_row = int(len(random_df) / 2)
    
    # Select the first half and set as training set.
    # Select the second half and set as test set.
    train_df = random_df.iloc[0:last_train_row]
    test_df = random_df.iloc[last_train_row:]
    
   
    k_rmses = {}
    
  
    # Fit model using k nearest neighbors where k = 5.
    knn = KNeighborsRegressor(n_neighbors=5)
    knn.fit(train_df[training_columns], train_df[target_column])

    # Make predictions using model.
    predicted_labels = knn.predict(test_df[training_columns])

    # Calculate and return RMSE.
    mse = mean_squared_error(test_df[target_column], predicted_labels)
    rmse = np.sqrt(mse)
        
    k_rmses[5] = rmse
    return k_rmses

```

## Hyperparameter value for 2,3,4,5 and 6 best features

```python
' Write a function that trains and test multivariate models using different k values for the top 4 best features '
def knn_train_test(training_columns, target_column, df):
    np.random.seed(1)
    
    # Randomize order of rows in data frame.
    shuffled_index = np.random.permutation(df.index)
    random_df = df.reindex(shuffled_index)

    # Divide number of rows in half and round.
    last_train_row = int(len(random_df) / 2)
    
    # Select the first half and set as training set.
    # Select the second half and set as test set.
    train_df = random_df.iloc[0:last_train_row]
    test_df = random_df.iloc[last_train_row:]
    
    k_values = [i for i in range(1, 25)]
    k_rmses = {}
    
    for k in k_values:
        # Fit model using k nearest neighbors.
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_df[training_columns], train_df[target_column])

        # Make predictions using model.
        predicted_labels = knn.predict(test_df[training_columns])

        # Calculate and return RMSE.
        mse = mean_squared_error(test_df[target_column], predicted_labels)
        rmse = np.sqrt(mse)
        
        k_rmses[k] = rmse
    return k_rmses

# Store the RMSE values for the top 3 models i.e 2 best features, 3 best 
# features and 4 best features for 1 < k < 25
k_rmse_results = {}

for i in range(2,6):
    k_rmse_results['{} best features'.format(i)] = knn_train_test(
        sorted_features[:i],
        'price',
        numeric_cars
    )
k_rmse_results
```

## Cross Validation Technique for 2,3,4,5 and 6 best features where Number of Folds is 5 and the Number of Nearest Neighbors is 5

```python
from sklearn.model_selection import cross_val_score, KFold
```

```python
# Modified function for different k values
def knn_train_test(training_columns, target_column, df):
    np.random.seed(1)
    
    # Randomize order of rows in data frame.
    shuffled_index = np.random.permutation(df.index)
    random_df = df.reindex(shuffled_index)
    
    kf = KFold(n_splits = 5, shuffle=True, random_state=1)
    knn = KNeighborsRegressor()

    mses = cross_val_score(knn, X = random_df[training_columns],
                           y= random_df[target_column],
                           scoring = "neg_mean_squared_error",
                           cv = kf)
    rmses = np.sqrt(np.absolute(mses))
    avg_rmse = np.mean(rmses)

    return avg_rmse

# Store the RMSE values for the top 4 models i.e 2 best features, 3 best 
# features and 4 best features 
k_rmse_results = {}

for i in range(2,7):
    k_rmse_results['{} best features'.format(i)] = knn_train_test(
        sorted_features[:i],
        'price',
        numeric_cars
    )
k_rmse_results
```

**Conclusion**: We have seen improvement in RMSE values for 2,3,4,5 and 6 best features using cross validation technique where number of folds is 5 and the number of nearest neighbors is 5 (the default k in KNeighborsRegressor())
